# K-Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
ykm = km.fit_predict(x)
ykm
```

```python
km.cluster_centers_
```

# Elbow graph

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,21), wcss)
plt.xlabel('K-values')
plt.ylabel('WCSS')
```

# Final Model

```python
km = KMeans(n_clusters=5)
ykm = km.fit_predict(x)
ykm
```

```python
clusters = pd.DataFrame(ykm,columns=['Clusters'])
```

# hyperparameter tuning
- finding out the best eps values(k-dist)

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=7).fit(x)
dist,ind = neigh.kneighbors(x)
```

```python
pd.DataFrame(dist)
```

```python
distance = np.sort(dist[:,1],axis=0)
```

```python
distance
```

```python
sns.set_style('darkgrid')
plt.plot(distance)
```

```python
db = DBSCAN(eps=3,min_samples=5)
ydb = db.fit_predict(x)
ydb
```

# Evaluation of DBscan

```python
silhouette_score(x,ydb)
```

```python
cl = pd.DataFrame(ydb,columns=['Cluster'])
```

```python
airDB = pd.concat([air,cl],axis = 1)
airDB
```
